ninestein benchmarks: Intel Xeon E5520 CPU @ 2.27 GHz

```python
from matplotlib import cm
import re

Ns = [25000, 50000, 100000]

cmap = cm.get_cmap('gist_rainbow')

time = []

for Ns1 in Ns:
    time1 = []
    methods = []
    left = []
    color = []
    left1 = 0
    i = 0
    j = 0
    n = 0
    for line in open("../measurements/benchmark/benchmark-ninestein.dat"):
        g = re.match("(.*\))\s+(\d+)\s+([0-9.]+)", line)
        if not g:
            print "corrupted:", line
            continue 
            
        method = g.group(1)
        Ns0 = int(g.group(2))
        time0 = float(g.group(3))
            
        f = line.split()
        
        if Ns0 == Ns1:
            m = method
            m = m.replace('(', ' ').replace(')', '')
            m = m.replace('EnergyDetector', 'ED')
            m = m.replace('Detector', '')
            m = re.sub("Np=(\d+), L=\d+", r"$N'=\1$", m)
            m = re.sub("L=(\d+)", r"$L=\1$", m)
            
            c = cmap(i/7. + j*0.007)
            #color.append(c)
            color.append(.85)
            
            methods.append(m)
            
            time1.append(time0)
            
            left.append(left1)
            left1 += 1
            j += 1
            
            if n == 0 or n % 4 == 2:
                left1 += .9
                i += 1
                j = 0
            n += 1
            
    time.append(time1)
    
left = array(left)
time = array(time)
```

```python
def plot_time(i):
    height = time[i,:]/time[i,0]
    #height[1] = .01
    #height[-2] = 1
    
    figure(figsize=(13, 5))

    bottom = 1
    bar(left, height-bottom, color=color, bottom=bottom)
    xticks(left+.4, methods, rotation=60, horizontalalignment='right')
    xlim(-.5, left1+.5)
    gca().set_yscale('log')
    ylim(1e0, 1e6)
    
    for x, y in zip(left, height):
        text(x+.4, y*5., "%.0f" % y, horizontalalignment="center", rotation=70, backgroundcolor='w')
    
    ylabel("$t_{CPU}$ [relative to energy detector]")
    title("$t=25.0$ms ($N_s = %d$ksamples)" % (Ns[i]/1e3,))
    
    tight_layout()
    grid()
    
plot_time(0)
#savefig("time_comparison_25ms.png", dpi=120)
#savefig("time_comparison_25ms.eps")
```
