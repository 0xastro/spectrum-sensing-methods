```python
import re
import itertools
```

```python
ref_pinmin = {}

f = open("../pinmin/pinmin.dat")
for line in f:
    path, pinmin = line.split()
    pinmin = float(pinmin)
    
    if path.startswith("out/sim_micsoft_fs"):
        g = re.search("ks_(\w+(?:_l\d+)?)_\*", path)
        if g:
            det = g.group(1)
            
            ref_pinmin[det] = pinmin
```

```python
ref_pinmin
```

```python
def iter_campaigns():
    
    f = open("../pinmin/pinmin.dat")
    for line in f:
        path, pinmin = line.split()
        pinmin = float(pinmin)
        
        g = re.search("sim_micsoft_spurious_([0-9]+)dbm_([0-9-]+)khz_.*ks_(\w+(?:_l\d+)?)_\*", path)
        if g:
            fn = float(g.group(1))*1e3
            pn = float(g.group(2))
            det = g.group(3)
            
            yield fn, pn, det, pinmin
```

```python
def plot_for_det(det_name):
    l = sorted(iter_campaigns())

    a = []
    fnl = []
    for fn, g in itertools.groupby(l, key=lambda x:x[0]):

        fnl.append(fn)
        b = []
        for fn, pn, det, pinmin in g:
            if det == det_name:
                b.append((pn, pinmin))
        a.append(b)

    a = array(a)
    #figure(figsize=(10,8))
    
    
    for i, fn in enumerate(fnl):
        plot(a[i,:,0], a[i,:,1], 'o-', label="%.0f kHz" % (fn/1e3,))
        
    ref = ref_pinmin[det_name]
    a = axis()
    plot([a[0], a[1]], [ref, ref], 'k--')
    

    xlabel("Spurious power ; $P_n$ [dBm]")
    ylabel("Signal power @ $P_d = 0.9$ dBm ; $P_{in-min}$ [dBm]")
    axis([None,None,-120,-110])
    legend(loc="upper left")
    
    det_name_nice = det_name.upper().replace("_L", " L=")
    title("Effect of spurious on %s" % det_name_nice)
    grid()
```

```python
plot_for_det("cav_l15")
```

```python
plot_for_det("mac_l15")
```

```python
plot_for_det("ed")
```
